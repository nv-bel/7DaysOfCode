# Dados de empréstimos dos acervos do sistema de bibliotecas da UFRN.

Um dos objetivos de uma biblioteca é garantir que os materiais informacionais estejam sendo utilizados. Os empréstimos realizados podem ser um indicador, mesmo que de forma básica (pois você não consegue garantir que haja uma leitura ou utilização real).

Por este motivo, entender a quantidade de empréstimos se torna importante.

Questões de diferentes perspectivas podem surgir como:

- A quantidade de empréstimos está aumentando ou diminuindo ao decorrer dos últimos anos?
- Em quais bibliotecas do sistema estão a maior quantidade de empréstimos?
- Quais são os temas mais emprestados? E os menos?

```python
import pandas as pd
```

```python
dados_2010_1 = pd.read_csv('./emprestimos/emprestimos-20101.csv')
dados_2010_2 = pd.read_csv('./emprestimos/emprestimos-20102.csv')
dados_2011_1 = pd.read_csv('./emprestimos/emprestimos-20111.csv')
dados_2011_2 = pd.read_csv('./emprestimos/emprestimos-20112.csv')
dados_2012_1 = pd.read_csv('./emprestimos/emprestimos-20121.csv')
dados_2012_2 = pd.read_csv('./emprestimos/emprestimos-20122.csv')
dados_2013_1 = pd.read_csv('./emprestimos/emprestimos-20131.csv')
dados_2013_2 = pd.read_csv('./emprestimos/emprestimos-20132.csv')
dados_2014_1 = pd.read_csv('./emprestimos/emprestimos-20141.csv')
dados_2014_2 = pd.read_csv('./emprestimos/emprestimos-20142.csv')
dados_2015_1 = pd.read_csv('./emprestimos/emprestimos-20151.csv')
dados_2015_2 = pd.read_csv('./emprestimos/emprestimos-20152.csv')
dados_2016_1 = pd.read_csv('./emprestimos/emprestimos-20161.csv')
dados_2016_2 = pd.read_csv('./emprestimos/emprestimos-20162.csv')
dados_2017_1 = pd.read_csv('./emprestimos/emprestimos-20171.csv')
dados_2017_2 = pd.read_csv('./emprestimos/emprestimos-20172.csv')
dados_2018_1 = pd.read_csv('./emprestimos/emprestimos-20181.csv')
dados_2018_2 = pd.read_csv('./emprestimos/emprestimos-20182.csv')
dados_2019_1 = pd.read_csv('./emprestimos/emprestimos-20191.csv')
dados_2019_2 = pd.read_csv('./emprestimos/emprestimos-20192.csv')
dados_2020_1 = pd.read_csv('./emprestimos/emprestimos-20201.csv')
```

```python
df_emprestimos = pd.concat([dados_2010_1, dados_2010_2, dados_2011_1, dados_2011_2,
                            dados_2012_1, dados_2012_2, dados_2013_1, dados_2013_2,
                            dados_2014_1, dados_2014_2, dados_2015_1, dados_2015_2,
                            dados_2016_1, dados_2016_2, dados_2017_1, dados_2017_2,
                            dados_2018_1, dados_2018_2, dados_2019_1, dados_2019_2, dados_2020_1])

df_emprestimos
```
