A coluna 'localização' representa a organização/numeração temática dos livros, de acordo com a Classificação Decimal Universal (CDU). Para isso, temos que:
- 000 a 099: Generalidades. Ciência e conhecimento.
- 100 a 199: Filosofia e psicologia.
- 200 a 299: Religião.
- 300 a 399: Ciências sociais.
- 400 a 499: Classe vaga. Provisoriamente não ocupada.
- 500 a 599: Matemática e ciências naturais.
- 600 a 699: Ciências aplicadas.
- 700 a 799: Belas artes.
- 800 a 899: Linguagem. Língua. Linguística.
- 900 a 999: Geografia. Biografia. História."

> Então, temos que criar uma coluna categórica com o label sendo respectivo à sua numeração

```python
bins = [0, 99, 199, 299, 399, 499, 599, 699, 799, 899, 999]
names = ['Generalidades. Ciência e conhecimento', 'Filosofia e psicologia',
         'Religião', 'Ciências sociais', 'Classe vaga. Provisoriamente não ocupada',
         'Matemática e ciências naturais', 'Ciências aplicadas', 'Belas artes',
         'Linguagem. Língua. Linguística', 'Geografia. Biografia. História']

df['categoria_cdu'] = pd.cut(df['localizacao'], bins, labels=names)

df
```
